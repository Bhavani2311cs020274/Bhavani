```python
from PIL import Image
import numpy as np

img = Image.open("50-Photos-of-Cute-Kittens-That-Will-Make-You-Melt_GettyImages-187144066_SSedit_FT.jpg")

gray_img = img.convert("L")
img_array = np.array(gray_img)

print(img_array)
```

```python
import matplotlib.pyplot as plt

plt.imshow(gray_img, cmap='gray')
plt.axis('off')
```

```python
complement_array = 255 - img_array
complement_img = Image.fromarray(complement_array)

plt.imshow(complement_img, cmap='gray')
plt.axis('off')
```

```python
plt.figure(figsize=(6,4))
plt.hist(img_array.flatten(), bins=256, range=(0,255), color='gray')
plt.title("Histogram of Original Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

```python
min_pixel = np.min(img_array)
max_pixel = np.max(img_array)

contrast_img = (img_array - min_pixel) * 255 / (max_pixel - min_pixel)
contrast_img = contrast_img.astype(np.uint8)
```

```python
plt.imshow(contrast_img, cmap='gray')
plt.title("Contrast Enhanced Image")
plt.axis('off')
```

```python
Image.fromarray(contrast_img).save("contrast_enhanced_image.jpeg")

```
